```python
import numpy as np
from sklearn.linear_model import LinearRegression
import csv
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
with open("/home/gato/Scripts/py_ex/LearnPython-main/files/cars.csv", "r") as f:
    csv_reader = csv.DictReader(f)
    cars = list(csv_reader)
```

```python
horsepower = [int(keys['Horsepower']) for keys in cars]
torque = [int(keys['Torque']) for keys in cars]
x = np.array(horsepower).reshape(-1,1)
y = np.array(torque).reshape(-1,1)
```

```python
l_r = LinearRegression()
l_r.fit(x, y)
y_pred = l_r.predict(x)
```

```python
##
## the interceptor and the slope
alpha = str(round(l_r.intercept_[0], 5)) ## el interceptor[0] y cinco decimales
beta = str(round(l_r.coef_[0][0], 5))
alpha, beta
```

```python
fig1, ax1 = plt.subplots()
ax1.set_title('interceptor: {} and pendiente: {}'.format(alpha, beta))
ax1.scatter(x, y)
ax1.plot(x, y_pred, c='r', alpha=0.25)
```
